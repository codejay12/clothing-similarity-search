```python
url="https://www.similarweb.com/top-websites/lifestyle/fashion-and-apparel/"
response = requests.request("GET", url, headers=headers)

# parse the downloaded data
soup = BeautifulSoup(response.text, 'html.parser')
productlist = soup.find_all("tr",{"class":"tw-table__row"})

```

```python
productlinks = []
for product in productlist:
        link = product.find("a",{"class":"tw-table__row-compare"}).get('href')
        productlinks.append("www."+(link.split("/website/")[1]).split("/")[0])

print(productlinks)
```

```python
url="https://www.amazon.in/s?k=men+shirts&crid=R6UJ0QNOMEDR&sprefix=men+shirt%2Caps%2C374&ref=nb_sb_noss_1"
response = requests.request("GET", url, headers=headers)

# parse the downloaded data
soup = BeautifulSoup(response.text, 'html.parser')
products= soup.find_all("div",{"class":"a-section a-spacing-base a-text-center"})
print(products)
```

```python
product_links = []
for product in products:
        link = product.find("a",{"class":"a-link-normal s-no-outline"}).get('href')
        product_links.append("https://www.amazon.in"+link)
```

```python
data=[]
for link in (product_links[0]):
    f = requests.get(link,headers=headers).text
    hun=BeautifulSoup(f,'html.parser')

    try:
        price=hun.find("div",{"class":"product-intro__head-price j-expose__product-intro__head-price"}).text.replace('\n',"")
    except:
        price = None

    try:
        about=hun.find("div",{"class":"product-intro__description"}).text.replace('\n',"")
    except:
        about=None

    try:
        rating = hun.find("div",{"class":"rate-num-small"}).text.replace('\n',"")
    except:
        rating=None

    try:
        name=hun.find("h1",{"class":"product-intro__head-name"}).text.replace('\n',"")
    except:
        name=None

    whisky = {"name":name,"price":price,"rating":rating,"about":about}

    data.append(whisky)
```

```python
#Loading the required libraries
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#read the file
data=pd.read_csv("main_data.csv")
```

```python
data
```

```python
new=[]
for i in data["title"]:
  i=re.sub('[\s+\d+:\.\)\( ]',' ',i) #'\.' spaces, numbers, colon, paranthesis, full stop rmoval
  i=re.sub(r'\S*@\S*\s?','',i)  # Email removal
  new.append(i)

data["title"]=new

```

```python
data=data.drop_duplicates()
data=data.dropna(axis=0) #drop duplicates
```

```python
data.isnull().sum()
```

#Approach-1

```python
##convert a words
tfidf=TfidfVectorizer()
mat= tfidf.fit_transform(data["title"]) 
```

```python
df=pd.DataFrame(mat.todense(),columns=tfidf.get_feature_names_out())
df=pd.concat([pd.Series(data["url"]),df],axis=1)
```

```python
df=df.dropna(axis=0)
```

```python
df
```

```python
def top_n_get_results(test,n):
    t1=tfidf.transform([test])
    sim1=cosine_similarity(np.asarray(t1.todense()),np.asarray(mat.todense())).reshape(-1)
    res=pd.concat([df["url"],pd.Series(sim1).rename("score")],axis=1)
    res=res.sort_values(by=["score"],ascending=False)[0:n]
    top_n=np.array(res["url"])
    results={"urls":top_n}
    return results
    
top_n_get_results("men black white shirt",5)
```
